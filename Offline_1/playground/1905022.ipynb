# Task B

```python
dataframe.dropna(subset=["Attrition"],inplace=True)
dataframe.fillna(dataframe.mean,inplace=True)
dataframe.drop_duplicates(inplace=True)
```

# Task C

```python
features=dataframe.drop(columns=["Attrition"],axis=1)
labels=dataframe["Attrition"]
```

# Task D

```python
from sklearn.preprocessing import LabelEncoder


features_encoded = pd.get_dummies(features, drop_first=True)
le=LabelEncoder()
features_encoded=features_encoded.apply(le.fit_transform)

print("Final encoded features DataFrame:")

features_encoded
```

```python

labelencoder=LabelEncoder()
label_encoded=labelencoder.fit_transform(labels)

label_encoded
```

# Task E

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def getScaledData(scaler, data):
    scaledData = scaler.fit_transform(data)
    return scaledData
```

```python
scaler=getScaledData(MinMaxScaler(),features_encoded)
scaler
```

```python
one_hot_cols=[]

for col in features_encoded.columns:
    if features_encoded[col].nunique() <=2:
        one_hot_cols.append(col)

one_hot_data=features_encoded[one_hot_cols]
numeric_data=features_encoded.drop(columns=one_hot_cols)

# one_hot_data
numeric_data
```

```python
# scaler=getScaledData(MinMaxScaler(),numeric_data)
scaler=getScaledData(StandardScaler(),numeric_data)
scaler=pd.concat([pd.DataFrame(scaler,columns=numeric_data.columns),one_hot_data],axis=1)
```

# Task F

```python
features_df = pd.DataFrame(scaler) # scaled feature dataframe
label_df = pd.DataFrame(label_encoded, columns=['Attrition'])

# features_df
# label_df
target_series = label_df['Attrition']
target_series
correlations = features_df.corrwith(target_series)
# # correlations

# top_20_correlations = correlations.abs().sort_values(ascending=False).head(20)
top_20_correlations = correlations.reindex(correlations.abs().sort_values(ascending=False).index).head(20)


top_20_correlations
```

# Plot

```python
import matplotlib.pyplot as plt
import numpy as np

top_20_features = top_20_correlations.index

unique_classes = label_df["Attrition"].unique()


for feature in top_20_features:
   
   
    for class_label in unique_classes:
        class_data = features_df.loc[label_df["Attrition"] == class_label]
        if class_label%2==1:
            plt.plot(class_data[feature], np.zeros_like(class_data[feature]), '+', label=f'Class {class_label}')
        else:
            plt.plot(class_data[feature], np.zeros_like(class_data[feature]), '-', label=f'Class {class_label}')
    
    # Customize the plot
    plt.legend()
    plt.title(f'1D Scatter Plot of {feature} by Numeric Classes')
    plt.xlabel(feature)
    plt.show()
```
